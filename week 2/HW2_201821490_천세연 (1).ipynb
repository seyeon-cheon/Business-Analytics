# HW2 소셜 네트워크 분석



- 네트워크: 각각의 객체들이 상호 연결된 구조
- 구성요소: node(vertex)와 link(edge)
- node: 고유한 속성을 가지고 있는 행위자
- 링크: 노드들간의 관계

즉, 네트워크는 복수의 노드들이 연결된 구조

#### 바베라스와 리빗의 실험
- 그룹 활동에 참여하고 있는 구성원들의 의사소통 패탄과 효과에 대한 실험을 수행. 관계에는 패턴이 있지 않을까?
    
- 실험 결과는 중심이 있는 스타형과 Y형 구조가 문제를 가장 빨리 해결하고, 의사전달 과정에서 오류도 가장 적음

- 네트워크 중심성에 대한 연구를 촉발했음

#### 프리만의 '중심성 지표' 개발

- 연결정도 중심성, 근접 중심성, 매개 중심성
- 매개 중심성은 중개자 혹은 매개자 역할을 얼마나 수행하는지 측정하는 지표

#### 버트의 '구조적 공백'

- 구조적 공백의 위치에 있는 노드는 다른 노드들과 중복관계를 가지고 있지 않기 때문에 노드와 노드를 연결 시켜주는 역할을 함. 

- 이는 그라노베티의 '약한 연결' 개념을 더 구체화 한 것으로 볼 수 있다. 
- 약한 연결 개념은 직업을 구하는 경로가 가까운 사람이 아닌 가볍게 알고 지내는 지인을 통해서의 경우가 많은 것이다.  

- 실제 조직 내에서 구조적 공백이 많은 사람이 적은 사람의 경우보다 더 빨리 승진을 한다

#### 왓츠, 스트로가츠의 '좁은 세상 네트워크'

- 인접한 노드와 일정한 규칙에 따라 일정한 숫자로만 연결 되어있는 정규 네트워크(Regular network), 무작위로 연결 되어있는 무작이 네트워크(Random network), 중간에 존재하는 좁은 세상 네트워크(small world network)를 제시
- 무척도 네트워크(scale-free network)가 밝혀지는 계기 제공
    - 분포가 정규분포를 따르지 않고 멱함수법칙을 따른다.
    - 소수의 node가 전체 네트워크에서 Hub 등 중요한 역할을 한다


## 그래프의 기본 개념

### 그래프 기본 개념
- 객체들 간의 네트워크 관계 모델을 수학적으로 펴현된 그래프로 나타내는 연구분야 그래프는 노드로 구성. 노드들은 링크로 연결되어있음. 

#### 네트워크

- 무방향 네트워크(Undirected): 링크의 방향이 없음. 책, 논문 공동 저자, 페이스북 친구 관계 등
- 방향 네트워크(Directed): 인터넷 주소의 하이퍼링크, 전화통화, 트위터의 리트윗 등
- 이진 네트워크(binary network): 관계의 유무만을 표현함
- 가중 네트워크(valued network): 관계의 정도를 함께 표현함

### 그래프 표현 방법

#### 인접 행렬(Adjaceny matrix)

- 인접한 노드들 간의 연결관계를 행렬로 표현,
- 네트워크 상에서 노드의 개수가 n일 경우, 인접 행렬은 nxn의 정방 행렬로 표현. 
- 행과 열이 교차하는 지점의 값은 관계 유무를 0과1로 표현함.
- 무방향 네트워크는 모든 관계가 쌍방형이어서 행렬이 대칭적이라는 특징을 지님.

#### 노드 리스트 (Node list)

- 인접한 링크가 있는 노드들만 표기한 방법

#### 엣지리스트 (Edge list)

- 노드들을 연결한 링크들의 리스트로 "링크의 목록"을 나타냄
- 고립된 노드의 경우 링크가 존재하지 않으므로 목록에 나타나지 않음 


### 네트워크 유형

#### 방향성과 가중치에 따른 분류 
- 방향성 유무와 가중치 유무에 따라 4가지 유형으로 구분 가능

#### 관계 위상에 따른 분류 

#### 네트워크 크기에 따른 분류
- 노드, 링크의 규모 등으로 분류함. 

#### 분석 대상에 따른 분류
- 전체 네트워크(complete network): 네트워크를 구성하고 있는 모든 노드를 포함한 네트워크를 말함. 
- 하위 네트워크(sub network): 전체 네트워크의 일부분. 부속되어있는 것으로 컴포넌트, 파당, 클러스트 등이 해당 (p31)



## 소셜 네트워크 분석 지표

### 네트워크 수준의 속성 분석

#### 네트워크 크기

- 네트워크를 구성하는 노두 수
- 네트워크 크기가 증가할수록 네트워크 구조의 복잡성도 증가
- 방향성이 있는 경우: nx(n-1)
- 방향성이 없는 경우: nx(n-1)/2

#### 네트워크 밀도 

- 네트워크 구성하는 노드들간의 연결 정도
- 노드간의 연결이 많아질수록 밀도는 높아진다
- 네트워크밀도 계산: k= 실제 연결된 링크 수, n= 노드 수
    - 무방향 이진 네트워크: k/{nx(n-1)/2}
    - 방향 이진 네트워크: k/{nx(n-1)}

#### 포괄성 (Inclusiveness)
- 전체 네트워크의 밀집도를 연결된 노드의 관점에서 표현
- 네트워크상 총 노드의 개수중 실제로 연결 관계가 존재하는 노드의 개수가 차지하는 비율

- 포괄성 계산
    - n-n(isolated)/n
    
#### 연결정도 
- n개의 노드가 존재한다고 가정했을 때, 각각의 노드에 최대 n-1개까지의 연결 정도가 가능
- 고립된 노드의 경우에는 연결정도가 0
- 연결정도가 높을수록 다른 노드들과 많은 관계를 가지고 있음 = 높은 활동성과 영향력을 반증


#### 연결 강도
- 노드간의 링크 강도
- 이진, 가중 네트워크
- 가중 나테워크에서는 상대적 가중치를 함께 표시 가능

#### 연결 거리
- 두개의  특정 노드들 사이에 존재하는 다양한 연결경로중, 가장 적은 수의 링크 단계를 거치는 경로. 가장 짧은 연결 경로
- 연결 거리가 짧을수록 노드간의 연결이 높아짐


### 노드 수준의 속성 분석

#### 직경
- 네트워크 상의 모든 연결거리 중 가장 긴 연결거리
- 네트워크 상에서 가장 멀리 떨어져 있는 두 노드 연결 = 네트워크 지름
- 직경이 작을수록 신속한 정보 전달이 가능함

#### 평균 연결 거리
- 네트워크 내 모든 노드 쌍의 최단 연결거리를 산출하여 평균값을 계산한 것.


### 네트워크에 내재된 속성 분석

#### 상호성(Reciprocity)
- 방향 네트워크 상에서 2개 노드간 연결이 쌍방일 경우: 상호적 연결
- 단방향인 경우: 비상호적 연결

- 노드 쌍 기준 상호성
    - 상호적으로 연결된 노드 쌍/ 연결된 전체 노드 쌍
- 연결 관계 기준 상호성
    - 상호적으로 연결된 링크 수/ 연결된 전체 링크 수
    
#### 군집화 계수 (Clustering coefficient)
- 군집화: 네트워크 상에서 특정 부분의 노드들이 밀접하게 연결되어 다른 부분보다 네트워크 밀도가 높아지는 현상
- 지역 군집화 계수: k = 해당 노드의 이웃 노드 수
    - N(이웃 노드들간에 실제 존재 링크 수)/{kx(k-1)/2}

## 소셜 네트워크 분석 지표

### 중심성 지표
- 한 행위자가 전체 네트워크 중심에 위치하는 정도로 표현한 지표
- 전체 네트워크에서 중요한 역할을 하는 노드를 파악

#### 연결정도 중심성

- 네트워크 상에서 한 노드가 얼마나 많은 연결관계를 가지고 있느지를 측정
- 노드의 연결된 링크 수가 많아질수록 노드의 연결 정도 중심성은 높아짐

- 연결정도 중심성: 절대적-> 노드 연걸 관계 수, 상대적 -> D/n-1

#### 근접 중심성
- 네트워크 상의 노드들간의 근접도를 기준
- 해당 노드가 전체 네트워크에서 얼마나 중앙에 위치하는가
- 절대적 근접 중심성: 1/ 해당 노드와 나머지 모든 노드간의 연결거리 합
    - 분모가 작을수록, 즉 연결이 작을수록 중심성이 커진다
- 상대적 근접 중심성: C x (n-1)

#### 매개 중심성 
- 네트워크 상에서 특정 노드가 다른 노드들의 중간에서 얼마나 중개자 및 매개자 역할을 하고 있는지를 측정
- 매개 중심성이 높을수록 네트워크 내에서 정보 및 자원의 흐름에 있어 높은 통제력 확보

#### 아이겐벡터 중심성
- 해당 노드에 직접 연결된 다른 노드들의 개수 뿐만 아니라 연결된 다른 노드들의 중요도 또한 함께 반영한것

- 중심성이 높은 노드들과 많이 연결된 노드일수록 해당 노드의 아이겐벡터 중심성은 커짐
